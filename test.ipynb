```python
# 예시
name = '롤접고갓생살러감'
url = "https://kr.api.riotgames.com/lol/summoner/v4/summoners/by-name/" + quotekr(name)
summoner = requests.get(url, headers={"X-Riot-Token": api_key})

print("status_code: {}".format(summoner.status_code))
if summoner.status_code == 200:
    print("{}의 소환사 레벨: {}".format(summoner.json()["name"], summoner.json()["summonerLevel"]))
```
